## Transforming the WAP variables to 10^x

```python
test = trainset.WAP102
```

```python
test = pd.DataFrame(test)
```

```python
test.info()
```

```python
bla = pd.DataFrame({"col": [1,-2,3]})
bla["col"] = bla.col.apply(float)
bla["potencia"] = 10 ** bla["col"]
#print(potencia)
print(bla)
```

```python
test["WAP102"] = test.WAP102.apply(float)
test["exp_10"] = 10 ** test["WAP102"]
print(test.head())
```

### Replacing 1e+100 with zeros

```python
test.loc[test["WAP102"] == 100, ["exp_10"]] = 0
print(test.head())
```

## Cleaning the data

# Removing WAPs with only 100

```python
trainset.head()
```

```python
trainset.describe()
```

```python
trainset = trainset.loc[:, (trainset.min() != 100)]
```

```python
print(trainset.head())
```

### Replacing 100 with (-200)

```python
trainset.min()[:465].min()
```

```python
temp = trainset.iloc[:,:465].replace(100, -200)
```

```python
trainset.head()
```

```python
temp["LONGITUDE"] = trainset.LONGITUDE
```

```python
temp["LATITUDE"] = trainset.LATITUDE
```

```python
temp["FLOOR"] = trainset.FLOOR
```

```python
temp["BUILDINGID"] = trainset.BUILDINGID
```

```python
temp["SPACEID"] = trainset.SPACEID
```

```python
temp["RELATIVEPOSITION"] = trainset.RELATIVEPOSITION
```

```python
temp["USERID"] = trainset.USERID
```

```python
temp["PHONEID"] = trainset.PHONEID
```

```python
temp["TIMESTAMP"] = trainset.TIMESTAMP
```

```python
trainset = temp[:]
trainset.head()
```
